```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def authors_by_year(journal):
    authors = pd.read_csv(f"../author_data/cleaned/{journal}_authors_cleaned.csv")
    counts = authors['Year'].value_counts().sort_index()
    plt.plot(counts)
    plt.show()
```

```python
authors_by_year("aer")
```

```python
authors_by_year("eco")
```

```python
authors_by_year("jpe")
```
